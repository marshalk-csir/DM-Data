```python
# Read CSV File and Add Totals Column 
# crime_stats['Severity'] = crime_stats.sum(axis=1)

def crime_schedule(value):
    if value == 'All theft not mentioned elsewhere':
        return "Schedule 2 (Minor Offences)"
    if value == 'Theft out of or from motor vehicle':
        return "Schedule 2 (Minor Offences)"
    if value == 'Drug-related crime':
        return "Schedule 5 (Serious Offences)"
    if value == "Robbery with aggravating circumstances":
        return "Schedule 6 (Most Serious Offences)"
    if value == "Common assault":
        return "Schedule 1 (Minor Offences)"
    if value == "Burglary at residential premises":
        return "Schedule 2 (Minor Offences)"
    if value == "Assault with the intent to inflict grievous bodily harm":
        return "Schedule 6 (Most Serious Offences)"
    if value == "Theft of motor vehicle and motorcycle":
        return "Schedule 2 (Minor Offences)"
    if value == "Shoplifting":
        return "Schedule 2 (Minor Offences)"
    if value == "Malicious damage to property":
        return "Schedule 1 (Minor Offences)"
    if value == "Common robbery":
        return "Schedule 1 (Minor Offences)"
    if value == "Driving under the influence of alcohol or drugs":
        return "Schedule 3 (minor offences – fine)"
    if value == "Sexual Offences":
        return "Schedule 2 (Minor Offences)"
    if value == "Stock-theft":
        return "Schedule 2 (Minor Offences)"
    if value == "Attempted murder":
        return "Schedule 5 (Serious Offences)"
    if value == "Robbery at non-residential premises":
        return "Schedule 2 (Minor Offences)"
    if value == "Robbery at residential premises":
        return "Schedule 2 (Minor Offences)"
    if value == "Murder":
        return "Schedule 6 (Most Serious Offences)"
    if value == "Illegal possession of firearms and ammunition":
        return "Schedule 5 (Serious Offences)"
    if value == "Arson":
        return "Schedule 2 (Minor Offences)"
    if value == "Robbery of cash in transit":
        return "Schedule 6 (Most Serious Offences)"
    if value == "Sexual offences as result of police action":
        return "Schedule 2 (Minor Offences)"
    if value == "Truck hijacking":
        return "Schedule 6 (Most Serious Offences)"
    if value == "Carjacking":
        return "Schedule 6 (Most Serious Offences)"
    if value == "Commercial crime":
        return "Schedule 5 (Serious Offences)"
    else:
        return None
 
crime_stats['Crime Schedule'] = crime_stats['Category'].map(crime_schedule)
display(crime_stats.head(20))
```

#### Create new Crime Descriptions column with additional key words

```python
# Amend Crime Descriptions

def crime_desc(value):
    if value == 'All theft not mentioned elsewhere':
        return "All theft not mentioned elsewhere; theft (value exceeds R2 500);"
    if value == 'Theft out of or from motor vehicle':
        return "Theft out of or from motor vehicle; consists of the unlawful and intentional removal of parts, accessories or equipment, that form part of a motor vehicle, or articles in or on the vehicle"
    if value == 'Drug-related crime':
        return "Drug-related crime; crime to use, possess or possession, manufacture, or distribute drugs classified as having a potential for abuse. Cocaine, heroin, marijuana, and amphetamines are examples of drugs classified to have abuse potential. Drugs are also related to crime through the effects they have on the user’s behavior and by generating violence and other illegal activity in connection with drug trafficking."
    if value == "Robbery with aggravating circumstances":
        return "Robbery with aggravating circumstances; unlawful and intentional forceful removal and appropriation in aggravating circumstances of movable tangible property belonging to another. Robbery cases are included in this category if any weapon, not restricted to a firearm, was used to commit the crime."
    if value == "Common assault":
        return "Common assault; is the unlawful and intentional Direct and indirect application of force to the body of another person, or Threat of application of immediate personal violence to another, in circumstances in which the threatened person is prevailed upon to believe that the person who is threatening him has the intention and power to carry out his threat. The act may consist in the direct or indirect application of force or threats of force decisive factor is the use of force or violence"
    if value == "Burglary at residential premises":
        return "Burglary at residential premises; Also refered to as “housebreaking” is a crime committed by a person who unlawfully and intentionally breaks into a building or similar structure, used for human habitation, with the intention to commit a crime on the premises."
    if value == "Assault with the intent to inflict grievous bodily harm":
        return "Assault with the intent to inflict grievous bodily harm (GBH); is the unlawful and intentional direct or indirect application of force to the body of another person with the intention of causing grievous bodily harm to that person. Note: The victim does not need to sustain serious injuries. The decisive element is the intention to cause serious injuries."
    if value == "Theft of motor vehicle and motorcycle":
        return "Theft of motor vehicle and motorcycle; consists of the stealing of a motor vehicle or motorcycle belonging to another person."
    if value == "Shoplifting":
        return "Shoplifting; consists of stealing an article for sale from a self-service shop during trading hours."
    if value == "Malicious damage to property":
        return "Malicious damage to property; in the unlawful and intentional damaging of property belonging to another."
    if value == "Common robbery":
        return "Common robbery; is the unlawful and intentional forceful removal and appropriation of movable tangible property belonging to another. Note: The decisive factor is the use of force or violence. Incidents of robbery (including business and house robbery) are classified as common robbery if no weapon is used."
    if value == "Driving under the influence of alcohol or drugs":
        return "Driving under the influence of alcohol or drugs; Drunk-driving, drinking and driving, DUI; DUI, or driving under the influence, means that you are in control of a mode of transport on a public road with blood levels exceeding the legal alcohol limit. ‘mode of transport’ as a vehicle and this includes bicycles, forklifts and mule carts"
    if value == "Sexual Offences":
        return "Sexual Offences; This category of crime replaces the former categories of rape and indecent assault. The new definition is contained in the Criminal Law (Sexual Offences and Related Matters) Amendment Act, Act 32 of 2007. This broad crime category includes over 50 crime types relating to violence sexual crimes, sexual exploitation, sexual grooming, trafficking and pornography."
    if value == "Stock-theft":
        return "Stock-theft; consists of the stealing of stock or produce belonging to another person"
    if value == "Attempted murder":
        return "Attempted murder; consists in the commission of an unlawful act with the intention of killing another human being but which does not result in the death of that human being (Not dead or person did not die)"
    if value == "Robbery at non-residential premises":
        return "Robbery at non-residential premises; House robbery is the unlawful and intentional forceful removal and appropriationof property from the non-residential premises of another person"
    if value == "Robbery at residential premises":
        return "Robbery at residential premises; House robbery is the unlawful and intentional forceful removal and appropriation of property from the residential premises of another person."
    if value == "Murder":
        return "Murder; consists in the unlawful and intentional killing of another human being (Died)"
    if value == "Illegal possession of firearms and ammunition":
        return "Illegal possession of firearms and ammunition; Unlawful possession of a firearm or ammunition can simply be defined as; 'Any person who possesses a firearm or ammunition without the relevant authorisation, license or permit."
    if value == "Arson":
        return "Arson; is the unlawful and intentional damaging of an immovable structure which is suitable for human occupation or the storing of goods and which belongs to another, by setting fire to it with the intention to prejudice another."
    if value == "Robbery of cash in transit":
        return "Robbery of cash in transit; is the unlawful and intentional forceful removal and appropriation of money or containers for the conveyance of money, belonging to another while such money or containers for the conveyance of money are being transported by a security company on behalf of the owner thereof."
    if value == "Sexual offences as result of police action":
        return "Sexual offences as result of police action; This category of crime replaces the former categories of rape and indecent assault. The new definition is contained in the Criminal Law (Sexual Offences and Related Matters) Amendment Act, Act 32 of 2007. This broad crime category includes over 50 crime types relating to violence sexual crimes, sexual exploitation, sexual grooming, trafficking and pornography."
    if value == "Truck hijacking":
        return "Truck hijacking; Robbery of a truck is the unlawful and intentional forceful removal and appropriation of a truck (excluding a light delivery vehicle) belonging to another."
    if value == "Carjacking":
        return "Carjacking; Robbery of a motor vehicle or “hijacking” is the unlawful and intentional forceful removal and appropriation of a motor vehicle (excluding a truck and a light delivery vehicle) belonging to another."
    if value == "Commercial crime":
        return "Commercial crime; Business robbery is the unlawful and intentional forceful removal and appropriation of property from the business of another person."
    else:
        return None
 
crime_stats['Crime Description'] = crime_stats['Category'].map(crime_desc)
display(crime_stats.head(20))
```
